```python
import pandas as pd
```

```python
player_names = "player_names.csv"
player_names_df = pd.read_csv(player_names)
player_names_df.head()
```

```python
atbats = "atbats.csv"
atbats_df = pd.read_csv(atbats)
atbats_df.head()
```

```python
atbats_df = atbats_df.rename(columns = {"batter_id":"id"})
```

```python
atbats_df.head()
```

```python
games = "games.csv"
games_df = pd.read_csv(games)
games_df.head()
```

```python
merge_df = pd.merge(player_names_df, atbats_df, on = "id")
```
